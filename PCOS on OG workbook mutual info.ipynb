# Dividing Dataset (train and Test)

```python
#first splitting dataset into dependent and independent features
from sklearn.model_selection import train_test_split
X=df.drop(labels=['PCOS (Y/N)'], axis=1)
Y=df['PCOS (Y/N)']
# separate dataset into train and test
X_train, X_test, Y_train, Y_test = train_test_split(
    df.drop(labels=['PCOS (Y/N)'], axis=1),
    df['PCOS (Y/N)'],
    test_size=0.3,
    random_state=0)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
#creating a function for the model
def models (x_train,y_train):
    
    #logistic Regression
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression(random_state=0)
    log.fit(X_train,y_train)
    
    #decision trees
    from sklearn.tree import DecisionTreeClassifier
    tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
    tree.fit(X_train,y_train)
    
    #Random Forest Classifier
    from sklearn.ensemble import RandomForestClassifier
    forest=RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
    forest.fit(X_train,y_train)
    
    #SVM linear
    from sklearn import svm
    clf=svm.SVC(kernel='linear', probability=True)#linear kernel
    clf.fit(X_train,y_train)
    
    #SVM polynomial Kernel
    from sklearn.svm import SVC
    svclassifier = SVC(kernel='poly',probability=True, degree=8)
    svclassifier.fit(X_train, y_train)
    
    #SVM gaussian Kernel
    from sklearn.svm import SVC
    svclassifier1 = SVC(kernel='rbf', probability=True)
    svclassifier1.fit(X_train, y_train)
    
    #SVM Signmoid
    from sklearn.svm import SVC
    svclassifier2 = SVC(kernel='sigmoid', probability=True)
    svclassifier2.fit(X_train, y_train)
    
    #naive-bayes classification
    from sklearn.naive_bayes import GaussianNB
    gnb = GaussianNB()
    gnb.fit(X_train, y_train)
    
    #KNN algo
    from sklearn.neighbors import KNeighborsClassifier
    knn = KNeighborsClassifier(n_neighbors=20)
    knn.fit(X_train, y_train)
    
    #AdaBoost Classifier
    from sklearn.ensemble import AdaBoostClassifier
    ada=AdaBoostClassifier(random_state=96)
    ada.fit(X_train, y_train)
    
    #Gradient Boosting algo
    from sklearn.ensemble import GradientBoostingClassifier
    gbc=GradientBoostingClassifier(n_estimators=500,learning_rate=0.05,random_state=100,max_features=5 )
    gbc.fit(X_train,y_train)
    
    
    #XgBoost algo
    from xgboost import XGBClassifier
    xgb = XGBClassifier()
    xgb.fit(X_train, y_train)
    
    
    #Extra trees algo
    from sklearn.ensemble import ExtraTreesClassifier
    etc= ExtraTreesClassifier()
    etc.fit (X_train, y_train)
    
    #lightgbm algo
    from lightgbm import LGBMClassifier
    LGBMClassifier.fit(X_train, y_train)
   

    
    
    
    

    
    
    

    print('[0] Logistic Regression Training Accuracy:',log.score(X_train, y_train))
    print('[1] Decision Tree Classifier Training Accuracy:', tree.score(X_train,y_train))
    print('[2] Random Forest Classifier Training Accuracy:',forest.score(X_train,y_train))
    print('[3] Support Vector Machine (linear) Training Accuracy:',clf.score(X_train,y_train))
    print('[4] Support Vector Machine (Polynomial) Training Accuracy:',svclassifier.score(X_train,y_train))
    print('[5] Support Vector Machine (Gaussian) Training Accuracy:',svclassifier1.score(X_train,y_train))
    print('[6] Support Vector Machine (Sigmoid) Training Accuracy:',svclassifier2.score(X_train,y_train))
    print('[7] Naive Bayes classification Training Accuracy:',gnb.score(X_train,y_train))
    print('[8] k-nearest neighbor classification Training Accuracy:',knn.score(X_train, y_train))
    print('[9] Adaboost classification Training Accuracy:',ada.score(X_train, y_train))
    print('[10] Gradient Boost classification Training Accuracy:',gbc.score(X_train, y_train))
    print('[11] XGBoost classification Training Accuracy:',xgb.score(X_train, y_train))
    print('[12] Extra trees classification Training Accuracy:',etc.score(X_train, y_train))
    print('[13] Lightgbm classification Training Accuracy:',LGBMClassifier.score(X_train, y_train))
    return log, tree, forest, clf, svclassifier, svclassifier1, svclassifier2, gnb, knn, ada, gbc, xgb, etc, LGBMClassifier
```

```python
model=models(X_train,Y_train)
```

```python
#test model accuracy on test data on confusion matrix
from sklearn.metrics import confusion_matrix

for i in range (len(model)):
    cm= confusion_matrix(Y_test, model[i].predict(X_test))
    
    TP=cm[0][0]
    TN=cm[1][1]
    FN=cm[1][0]
    FP=cm[0][1]
    
    print('Model',i)
    
    print(cm)
    plt.figure(figsize=(10,6))
    sns.heatmap(cm,annot=True,fmt="d")
    sns.heatmap
    plt.xlabel('Actual values')
    plt.ylabel('Predicted values')
    print('Testing accuracy=', (TP+TN)/(TP+TN+FN+FP))
```

```python
#show another way to get metrics of the model
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

for i in range (len(model)):
    print('Model',i)
    print(classification_report(Y_test,model[i].predict(X_test)))
    print('Accuracy:',accuracy_score(Y_test,model[i].predict(X_test)))
    print('AUC:',roc_auc_score(Y_test,model[i].predict(X_test)))
    y_pred = model[i].predict(X_test)
    y_proba = model[i].predict_proba(X_test)[:,1]
    fpr, tpr, _ = roc_curve(Y_test,  y_proba)
    auc = round(roc_auc_score(Y_test, y_proba),2)
    plt.plot(fpr,tpr,color='xkcd:green',label='Model AUC :' +str(auc))
    ax = plt.axes()
    ax.set_facecolor('beige')
    plt.legend(loc=4)
    plt.grid(True)
    plt.plot([0, 1], [0, 1], color='blue', linestyle='-.')
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.show()
```

## Creating the ANN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU, ELU, ReLU
from tensorflow.keras.layers import Dropout
#PReLu not possible
```

```python
#Sequential ensures we are doing Forward and Bakward Propogation at once
#dense is to add nodes in the hidden layers any where
#dropout to avoid overfitting
```

```python
#initializing the ANN
classifier= Sequential()

```

```python
## adding the input layer
#add all the inputs as the shape 
classifier.add(Dense(units= 43 , activation='relu'))
```

```python
#adding the First hidden layer
# maybe some of the original neurons not all
classifier.add(Dense(units=28 , activation='relu'))
classifier.add(Dropout(0.2))
```

```python
#adding the Second hidden layer
# maybe some of the original neurons not all
classifier.add(Dense(units=15, activation='relu'))
classifier.add(Dropout(0.3))
```

```python
#adding the third hidden layer
# maybe some of the original neurons not all
classifier.add(Dense(units=7, activation='relu'))
classifier.add(Dropout(0.2))
```

```python
#adding the fourth hidden layer
# maybe some of the original neurons not all
classifier.add(Dense(units=4 , activation='relu'))
classifier.add(Dropout(0.3))
```

```python
#adding the fifth hidden layer
# maybe some of the original neurons not all
classifier.add(Dense(units=3 , activation='relu'))
classifier.add(Dropout(0.2))
```

```python
#adding output
classifier.add(Dense(1, activation='sigmoid'))
```

```python
classifier.compile(optimizer ='adam', loss='binary_crossentropy',metrics=['accuracy'])
```
